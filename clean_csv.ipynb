```python
import numpy as np
import pandas as pd

def clean_data(df):
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    # Drop rows with missing data in column: 'Name'
    df = df.dropna(subset=['Name'])
    # Replace empty strings with NaN
    df['Magnitude'] = df['Magnitude'].replace('999.99', np.nan)
    df['Name'] = df['Name'].replace('             ', np.nan)
    df['Name'] = df['Name'].replace('	     ', np.nan)
    df['Name'] = df['Name'].replace('            ', np.nan)
    df['Name'] = df['Name'].replace('           ', np.nan)
    df['Name'] = df['Name'].replace('          ', np.nan)
    df['Name'] = df['Name'].replace('         ', np.nan)
    df['Name'] = df['Name'].replace('        ', np.nan)
    df['Name'] = df['Name'].replace('       ', np.nan)
    df['Name'] = df['Name'].replace('      ', np.nan)
    df['Name'] = df['Name'].replace('     ', np.nan)
    df['Name'] = df['Name'].replace('    ', np.nan)
    df['Name'] = df['Name'].replace('   ', np.nan)
    df['Name'] = df['Name'].replace('  ', np.nan)
    df['Name'] = df['Name'].replace(' ', np.nan)
    df['Name'] = df['Name'].replace('', np.nan)
    # Drop rows where column 'A' has NaN
    df = df.dropna()
    # Replace all instances of 0 with 0 in column: 'Magnitude'
    df.loc[df['Magnitude'] == 999.99, 'Magnitude'] = np.nan
    df=df.dropna()
    return df

# Loaded variable 'df' from URI: c:\Users\Mikae\source\repos\astro-script\ephe\star_names_and_magnitudes.csv
df = pd.read_csv(r'c:\\Users\\Mikae\\source\\repos\\astro-script\\ephe\\star_names_and_magnitudes.csv')

df_clean = clean_data(df.copy())
df_clean.head()
```
